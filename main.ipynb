### Modified for 224,224 images in color.

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array

face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
classifier = load_model('hybrid_vgg16_face_emotion_model-fulldataset.h5')

emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

cap = cv2.VideoCapture(0)  # Change to 1 if the default index (0) doesn't work

if not cap.isOpened():
    print("Error: Cannot open webcam.")
    # Add appropriate actions, e.g., return, exit the script, or display an error message

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Cannot read frame.")
        break

    # Detect faces
    faces = face_classifier.detectMultiScale(frame)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 255), 2)
        roi_color = frame[y:y + h, x:x + w]

        # Preprocess the image for the emotion classifier
        roi_color = cv2.resize(roi_color, (224, 224))
        roi_color = roi_color.astype('float') / 255.0
        roi_color = img_to_array(roi_color)
        roi_color = np.expand_dims(roi_color, axis=0)

        # Predict emotions
        prediction = classifier.predict(roi_color)[0]
        label = emotion_labels[prediction.argmax()]
        label_position = (x, y)
        cv2.putText(frame, label, label_position, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('Emotion Detector', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
